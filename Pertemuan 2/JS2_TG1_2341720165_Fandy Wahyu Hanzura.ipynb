Lakukan import library sesuai dengan kebutuhan

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

Lakukan load data

```python
dpath = '../data/wbc.csv'
df = pd.read_csv(dpath)
df.head()
```


1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
df.info()
```

Lakukan pemisahan untuk data Benign karena merupakan data yang dapat diidentifikasi karena merupakan diagnosa `jinak`

```python
# Memisahkan data 'jinak' (Benign)
data_benign = df[df['diagnosis'] == 'B'].copy()
data_benign.head()
```

Sedangka data malignant adalah data yang tidak dapat diidentifikasi karena merupakan diagnosa `ganas`

```python
data_malignant = df[df['diagnosis'] == 'M'].copy()
data_malignant.head()
```

2. Lakukan proses encoding pada kolom "diagnosis".

```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
df.head()
```


3. Lakukan proses standardisasi pada semua kolom yang memiliki nilai numerik.

```python
df_num = df.drop(columns = ['id', 'diagnosis'])
std = StandardScaler()
df = std.fit_transform(df_num)
```

Ubah kembali data kode ke dalam dataframe setelah sebelumnya menjadi numpy array

```python
framed = pd.DataFrame(df, columns = df_num.columns)
framed.head()
```
